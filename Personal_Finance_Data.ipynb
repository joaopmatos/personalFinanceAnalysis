```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

finance_data = pd.read_csv("/content/sample_data/personal_finance_tracker_dataset.csv", index_col="user_id")
```

Imported financial data

```python
subset = finance_data.head(500).copy()
# Relabel categories
subset["financial_scenario"] = subset["financial_scenario"].replace({
    "recession": "Recession Period",
    "inflation": "Inflation Period",
    "normal": "Stable Economy"
})
```

Reduced the data to a subset of 500 objects. Relabeled the categories.

```python
# Map the color to each hue level
custom_palette = {
    "High": "#1f77b4",   # strong blue
    "Medium": "#c4d7f0", # muted blue
    "Low": "#cbddea",    # light blue
}
hue_order = ["Low", "Medium", "High"]

scenario_order = ["Recession Period", "Inflation Period", "Stable Economy"]
```

Defined a custom pallete, the order of the hues and an order for the scenarios in which I'll present. Corresponds to the X axis

```python
sns.set_style("white")
sns.set_context("paper")

grid = sns.catplot(
    data=subset,
    x="financial_scenario",
    y="monthly_income",
    hue="financial_stress_level",
    kind="bar",
    order=scenario_order,
    hue_order=hue_order,
    palette=custom_palette
)

grid.set_axis_labels("Economic Scenario", "Monthly Income (USD)")
grid._legend.set_title("Financial Stress Level")
plt.title("Monthly Income Distribution by Scenario and Stress Level")

plt.show()
```

Generated a Bar Plot, including variability whiskers, helping us gauge the reliability of the measurement.

```python
grid = sns.catplot(
    data=subset,
    x="financial_scenario",
    y="monthly_income",
    hue="financial_stress_level",
    kind="box",
    order=scenario_order,
    hue_order=hue_order,
    palette=custom_palette
)
grid.set_axis_labels("Economic Scenario", "Monthly Income (USD)")
grid._legend.set_title("Financial Stress Level")
plt.title("Monthly Income Distribution by Scenario and Stress Level")
```
